```python
poverty_percent['population'] = poverty_percent['population'].astype(str).str.replace(',','')
poverty_percent['poverty'] = poverty_percent['poverty'].astype(str).str.replace(',','')
poverty_percent['population'] = poverty_percent['population'].astype(int)
poverty_percent['poverty'] = poverty_percent['poverty'].astype(int)
poverty_percent.head()
```

```python
poverty_percent['percent'] = (poverty_percent['poverty']/poverty_percent['population'])
```

```python
poverty_percent['percent'] = poverty_percent['percent'] * 100
poverty_percent
```

```python
poverty_percent['percent'] = poverty_percent['percent'].astype(int)
```

```python
for index, row in poverty_percent.iterrows():
    if row.percent <= 10 :
        poverty_percent.loc[index, 'poverty_bracket'] = 'minimal'
    elif row.percent >= 11 & row.percent <= 20:
        poverty_percent.loc[index, 'poverty_bracket'] = 'low'
    elif row.percent >= 21 & row.percent >= 30:
        poverty_percent.loc[index, 'poverty_bracket'] = 'moderate'
    elif row.percent >= 31 & row.percent >= 40:
        poverty_percent.loc[index, 'poverty_bracket'] = 'high'
    else :
        poverty_percent.loc[index, 'poverty_bracket'] = 'server'
    
poverty_percent.head()       
```
